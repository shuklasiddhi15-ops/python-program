## Question 8: Create a function that counts the number of vowels in a string.


```python
def vowels(word):
    vowels ="AEIOUaeiou"
    count =0
    for i in  word:
        if i in vowels:
            count += 1
            print(count)
vowels("education")
```

## Question 9: Create a function that returns the factorial of a given number.


```python
def fact_num(num):
    fact = 1
    for i in range (1,num+1):
        fact *= i
        print(fact)
fact_num(5)
```

```python
# Input: 5

# Output: 120
```

## Question 10: Create a function that checks whether a number is Prime or Not Prime.


```python
def prime_num(num):
    if num <= 1:
        return "not prime number"

    for i in range(2, num):
        if num % i == 0:
            return "not prime number"

    return "prime number"

print(prime_num(6))
```

```python
# Input: 13

# Output: Prime Number
```

## Question 11: Create a function that returns the largest element from a list.


```python
def largest_num():
    l = [12, 45, 98, 34, 21]
    max_val = l[0]
    for i in l:
        if i > max_val:
            max_val = i
    print(max_val)

largest_num()
```

## Question 13: Create a function that checks whether a string is a palindrome.


```python
def is_palindrome(word):
    reversed_word = word[::-1]

    if word == reversed_word:
        return True
    else:
        return False


word = input("Enter a string: ")

if is_palindrome(word):
    print("Palindrome")
else:
    print("Not a palindrome")
```

```python
# Input: madam

# Output: Palindrome
```

## Question 14: Create a function that returns the sum of digits of a number.


```python
def sum_of_digits(num):
    sum_digits = 0

    while num > 0:
        digit = num % 10
        sum_digits += digit
        num = num // 10

    return sum_digits

print(sum_of_digits(12345))
```

```python
# Input: 12345

# Output: 15
```

## Question 15: Create a function that counts uppercase and lowercase characters in a string.


```python
def count_case(text):
    upper_count = 0
    lower_count = 0

    for ch in text:
        if ch.isupper():
            upper_count += 1
        elif ch.islower():
            lower_count += 1

    print("Uppercase letters:", upper_count)
    print("Lowercase letters:", lower_count)

count_case("PyThOn")
```

```python
# Input: PyThOn

# Output:Uppercase: 3 Lowercase: 3
```
