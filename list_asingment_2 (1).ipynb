## Q1. Sort List in Ascending Order

Input
nums = [45, 12, 78, 23, 5]

Output
[5, 12, 23, 45, 78]

```python
l =[45,12,78,23,5]
for i in range (len(l)):
    for j in  range (i+1,len(l)):
        if l[i]>l[j]:
            l[i],l[j] = l[j],l[i]
print(l)
```

## Q2. Sort List in Descending Order

Input
nums = [45, 12, 78, 23, 5]

Output
[78, 45, 23, 12, 5]

```python
l =[45,12,78,23,5]
for i in range (len(l)):
    for j in  range (i+1,len(l)):
        if l[i]<l[j]:
            l[i],l[j] = l[j],l[i]
print(l)
```

## Q3. Find Largest Number

Input
nums = [12, 45, 89, 23, 67]

Output

```python
l = [12,45,89,23,67]
largest_num = l[0]
for i in l:
  if i > largest_num:
    largest_num = i
print(largest_num)
```

## Q5. Remove Duplicate Elements

Input
nums = [1, 2, 3, 2, 4, 1, 5]

Output
[1, 2, 3, 4, 5]

```python
l = [1,2,3,2,4,1,5]
unique_list =[]
for i in l:
  if i not in unique_list:
    unique_list.append(i)
print(unique_list)
```

## Q6. Count Frequency of Number 5

Input
nums = [5, 2, 5, 7, 5, 1]

Output
3

```python
l =[5,2,5,7,5,1]
count_of_5 = 0
for i in l:
  if i == 5:
    count_of_5 += 1
print(count_of_5)
```
