## Q2. Remove Duplicate Values

```
Input:
lst=[10,20,20,30,30,40]

Output:
{10,20,30,40}
```

````python
lst=[10,20,20,30,30,40]
unique_set = set()
for i in lst:
    unique_set.add(i)
print(unique_set)

````
