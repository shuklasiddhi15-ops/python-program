## Q2. Create an empty set and add numbers 5, 10, 15, 20, 25 using a loop. Print the set.

Output:
```
{5, 10, 15, 20, 25}
```

````python
s = {5,10,15,20,25}
s = set ()
for i in range (5,26,5):
 s.add(i)
print(s)
````
