## Question 1: Create a Dictionary

## Question 7: Count Total Keys

Input:
{'Name':'Raj','Age':20,'City':'Bhopal'}

Output:
3

```python
d = {'Name':'Raj','Age':20,'City':'Bhopal'}
count = 0
for i in d:
    count+=1
print(count)
```

## Question 14: Merge Two Dictionaries

Input:
d1={'a':1,'b':2}
d2={'c':3,'d':4}

Output:
{'a':1,'b':2,'c':3,'d':4}

```python
d1 = {'a':1,'b':2}
d2 = {'c':3,'d':4}
for i in d2:
  if i not in d1:
    d1[i] =d2[i]
print(d1)
```
